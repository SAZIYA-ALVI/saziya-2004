POSITIONAL ENCODING

Positional encoding is a method used in Transformer models to inject information about the order or position of tokens in a sequence, enabling the model to process sequential data without relying on recurrence or convolution.


```
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
input_token = tokenizer("Write a python code for adding two numbers", return_tensors="pt")
input_token
```

```
import torch
from transformers import AutoModelForCausalLM
model = AutoModelForCausalLM.from_pretrained("sentence-transformers/all-MiniLM-L6-v2",
torch_dtype = torch.bfloat16)
```

```
out = model(input_ids = input_token["input_ids"])
out
```

```
gen_out = model.generate(input_token["input_ids"])
gen_out
```

```
out = model(input_ids = input_token["input_ids"], max_new_tokens=100)
out
```

```
tokenizer.batch_decode(gen_out)
```
